### Colored Scatterplots

```python
iris = pd.read_csv(r"D:\ML\Machine-Learning\Datasets\Matplotlib-Dataset's\iris.csv")
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],
            c = iris['Species'],   # color by species - it assigns a color to each point according to its species. (0, 1, 2)
            cmap = 'plasma',          # colormap - it's a built-in parameter used to map numerical data values to a continuous or discrete gradient of colors
            alpha = 0.70)           # 70% opacity - controls the transparency level of your plot elements

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()  # adds legend bar
```

01) c=column
Maps values → colors

02) cmap='jet'
Color scheme. Try 'viridis', 'plasma', 'coolwarm'

03) alpha=0.7
Transparency. 0=invisible, 1=solid

04) plt.colorbar()
Shows the color scale legend

### Annotations

<p>Annotations let you label individual data points on a plot. Use plt.text(x, y, label) to place text at any coordinate. And also you can edit that label like font size and color using fonctdict = {'size' : 13, 'color' : 'green'}

```python
batters = pd.read_csv(r"D:\ML\Machine-Learning\Datasets\Matplotlib-Dataset's\batter.csv")
```

```python
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs']) # s = size of the points

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

### Horizontal and Vertical lines
<p>Reference lines mark thresholds or benchmarks on your chart. axhline draws a horizontal line across the full width; axvline draws a vertical line across the full height.

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs']) # s = size of the points

plt.axhline(130,color='red') # axis-horizontal line (strike rate = 130)
plt.axhline(140,color='green') # axis-horizontal line (strike rate = 140)
plt.axvline(30,color='red') # axis-vertical line (average = 30)

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

### Subplots
<p>Subplots let you place multiple charts in a single figure. There are two ways to do this — OOP style (recommended) and the older add_subplot style.


<p> Use Case: When into a large dataset, you're dealing with multiple columns and want to know it's relationship then we use the concept of "Subplots".


```python
# A diff way to plot graphs
batters.head()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

### A different way to plot this graph:
<p> From the subplots class we get two object of that class fig, and ax.

<p> Fig object is basically a graph where you plot the points
<p> Axis object (x, and y) there are both axis where it plotted.

```python
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
# batter dataset
```

<p> sharex = True means that both the graphs will have the same x-axis.

<p> Note that: If the axis is not same then you will never do True to the sharex or sharey

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6)) # sharex = True means that both the graphs will have the same x-axis.

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

### 3rd Way:

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1) # row, column, graph_number
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])
```

### 3D Scatter Plots
<p>Matplotlib supports 3D plots via the mpl_toolkits.mplot3d toolkit (auto-imported). You activate 3D by setting projection='3d' on your subplot.

```python
batters

fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(batters['runs'],batters['avg'],batters['strike_rate'],marker='+')
ax.set_title('IPL batsman analysis')

ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('SR')
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# 3D Surface Plot - Any ruff idea 
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
