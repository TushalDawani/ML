# **Importing Libraries:**

```python
# Importing library for LinearRegression
from sklearn.linear_model import LinearRegression

# Import library for ScatterPlot or Plotting Graphs:
import matplotlib.pyplot as plt
```

# **Scenario # 01: "Predicting house price based on house size"**

```python
 # Training data: house size (100 sq ft) → price (lakhs)
X = [[10], [15], [20], [25], [30]]   # Input features
y = [50, 75, 100, 125, 150]           # Target values

model = LinearRegression()            # Create model
model.fit(X, y)                       # TRAIN — model learns here

prediction = model.predict([[17]])    # Predict on new data
print(f'Predicted price: Rs {prediction[0]:.0f} Lakhs')
# Output: Predicted price: Rs 85 Lakhs
```

## **Scenario # 01: "Plotting"**

```python
# Ploting the price:
plt.scatter(X, y)
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price (lakhs)')
plt.title('House Size vs Price')

plt.show()
```

# **Scenario # 02: "Predicting exam score based on study hours"**

```python
 # Training data: house size (100 sq ft) → price (lakhs)
X = [[2], [3], [4], [5], [6], [7], [8], [9]]   # Input features
y = [45, 50, 55, 60, 65, 70, 75, 80]           # Target values

model = LinearRegression()            # Create model
model.fit(X, y)                       # TRAIN — model learns here

prediction = model.predict([[6.5]])    # Predict on new data
print(f'Predicted score: {prediction[0]:.0f}')
# Output: Predicted score: 68
```

## **Scenario # 02: "Plotting"**

```python
# Ploting the score:
plt.scatter(X, y)
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Score')

plt.show()
```
