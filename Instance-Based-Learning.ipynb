# **Instance-Based Learning:**
---

The algorithm memorises the entire training dataset. To make a prediction, it finds the most similar stored examples and uses their labels.


## **Primary Algorithm: K-Nearest Neighbours (KNN)**

### **Euclidean Distance** = sqrt( SUM[(xi - yi)^2] ) for all features

image.png

**Same it is for other students...**

---


```python
from sklearn.neighbors import KNeighborsClassifier

# Input Data:
X = [
    [8,90],
    [7,85],
    [2,40],
    [3,50],
    [6,80]
]

# Output Data:
y = ["Pass","Pass","Fail","Fail","Pass"]

# Creating KNN Model:
knn = KNeighborsClassifier(n_neighbors=3)

# Train Model to the given data:
knn.fit(X, y)

# Prediciting without any equation
prediction = knn.predict([[5,75]])

print(prediction)
```
